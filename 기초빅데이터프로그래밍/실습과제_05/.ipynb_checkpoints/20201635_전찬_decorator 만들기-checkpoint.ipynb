```python
class verbose:
    def __init__(self, f):
        print("Initializing Verbose.")
        self.func = f
    
    def __call__(self, name):
        print("\nBegin", self.func.__name__)
        self.func(name)
        print("End", self.func.__name__)

@verbose
def my_function(name):
    print("hello, " + name + "!")

print("Program start")
my_function("Mickey")
my_function("Minnie")
my_function("Donald")
```
